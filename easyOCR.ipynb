```python
import easyocr
import os
from PIL import Image
from google.colab import drive

drive_base_path = "/content/drive/MyDrive/amzon_ml_hackathon"
input_folder = os.path.join(drive_base_path, "images") # Input folder in Drive
output_folder = os.path.join(drive_base_path, "extracted_text")  # Output folder in Drive

# Ensure Output Folder Exists
os.makedirs(output_folder, exist_ok=True)

# Initialize EasyOCR Reader (English)
reader = easyocr.Reader(['en'])
```

```python
import re
# Process Each Image in the Folder
for filename in os.listdir(drive_base_path):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):  # Process only images
        image_path = os.path.join(drive_base_path, filename)

        print(f"Processing: {filename}")
        results = reader.readtext(image_path)

        # Extract Detected Text
        extracted_text = "\n".join([res[1] for res in results])

        # Extract Image Number (e.g., image0_width.jpg → image0.txt)
        match = re.search(r'image(\d+)', filename)
        if match:
            image_number = match.group(0)  # e.g., 'image0'
        else:
            image_number = os.path.splitext(filename)[0]  # Default to filename without extension

        # Define Output Text File Path
        output_text_path = os.path.join(output_folder, f"{image_number}.txt")

        # Save Extracted Text
        with open(output_text_path, "w", encoding="utf-8") as f:
            f.write(extracted_text)

        print(f"✅ Text saved to {output_text_path}")

print("All images processed successfully!")
```

```python
# Main code

import pandas as pd

# mapping the extracted text with the entity_name
results_folder = os.path.join(drive_base_path, "output")  # Folder for final CSV & results

'''# Ensure Results Folder Exists
os.makedirs(results_folder, exist_ok=True)'''

# Load Original CSV
csv_path = os.path.join(drive_base_path, "sample_test.csv")  # Update with actual CSV name
df = pd.read_csv(csv_path)

# Extract image<number> from filenames
def extract_image_number(filename):
    match = re.search(r'image(\d+)', filename)  # Extract image<number>
    return match.group(0) if match else None

# Create Mapping of Extracted Text
extracted_texts = {}
for file in os.listdir(output_folder):
    if file.endswith(".txt"):
        image_number = extract_image_number(file)  # Get image<number> from filename
        if image_number:
            with open(os.path.join(output_folder, file), "r", encoding="utf-8") as f:
                extracted_texts[image_number] = f.read().strip()  # Store extracted text

# Add Extracted Values to DataFrame
df['image_number'] = df['image_link'].apply(extract_image_number)  # Extract image<number> from CSV
df['extracted_value'] = df['image_number'].map(extracted_texts)  # Map extracted text

# Save Updated CSV to Results Folder
updated_csv_path = os.path.join(results_folder, "results.csv")
df.to_csv(updated_csv_path, index=False)

print(f"✅ Updated CSV saved to: {updated_csv_path}")
```
